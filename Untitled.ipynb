```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

internet = pd.read_csv("List of Countries by number of Internet Users - Sheet1.csv")
internet.shape
internet.columns
internet.isnull().sum()
internet.head()
data=internet[internet["Rank.i"]!= "-"].sort_values(by=["Population"],ascending=False)
data.head(20)
#Fig 1
x_country=data["Country or region"].head(20)
y_percent=(data["Internet Users"]/data["Population"]).head(20)
plt.title("Internet and the Population")
sns.barplot(x_country,y_percent)
plt.xticks(rotation=90)
plt.xlabel("Top 20 Most Populated countries")
plt.ylabel("Percentage of Non-Internet Users")
plt.show()
#Fig 1 ends
happiness_data=pd.read_csv("2019.csv")
happiness_data.columns
happiness_data.shape
data.shape
data.sort_values(by=["Population"])
result = pd.merge(data,happiness_data,on="Country or region")#merging two databases
result.head()
result.isnull().sum()
#Fig 2 
wh1=result[["Population","Internet Users","Score","GDP per capita","Healthy life expectancy","Generosity","Perceptions of corruption"]]
cor = wh1.corr() #Calculate the correlation of the above variables
sns.heatmap(cor, square = True) #Plot the correlation as heat map
#Fig 2 ends
#Fig 3
plt.scatter(result["Percentage"],result["GDP per capita"],s=np.array(result["Population"]/1000000)*2,alpha=0.7)
plt.xscale("linear")
plt.xlabel("Percentage of Internet Users")
plt.ylabel("GDP")
plt.title("Internet and GDP")
#Fig 4
plt.scatter(result["Percentage"],result["Healthy life expectancy"],s=np.array(result["Population"]/1000000)*2,alpha=0.7)
plt.xscale("linear")
plt.xlabel("Percentage of Internet Users")
plt.ylabel("Life Expectancy")
plt.title("Internet and Life Expectancy")
#Fig 5
plt.plot(result["Percentage"],result["Score"])
plt.xscale("linear")
plt.xlabel("Percentage of Internet Users(1-10)")
plt.ylabel("Happiness Ratings")
plt.title("Internet and Happiness")
#Fig 6
plt.plot(result["Percentage"],result["Perceptions of corruption"])
plt.xscale("linear")
plt.xlabel("Percentage of Internet Users(1-10)")
plt.ylabel("Perception of Corruption")
plt.title("Internet and Corruption Perception")
```
