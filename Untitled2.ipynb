```python
import numpy as np
import random 

def new_array(c,r):
    x = np.random.randint(10, size=(c, r))
    return x

def final_array(arr,ar):
    a,b= arr.shape
    for i in range(a):
        for j in range(b):
            for x in [-1,0,1]:
                for y in [-1,0,1]:
                    if (x==0 and y==0):
                        continue
                    x1=x+i
                    y1=y+j
                    if (x1<0) or (y1<0):
                        continue
                    try:
                        n1=arr[(x1,y1)]
                    except IndexError:
                        n1=0
                    ar[(i,j)]+=n1
    return ar

c = int(input("Columns : "))
r = int(input("Rows : "))
first_array = new_array(c,r)
ar=first_array*0
print(first_array)
f=final_array(first_array,ar)
print(f)
```
